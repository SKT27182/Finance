## Federal Funds Rate and EONIA

The Federal Funds Rate is the interest rate at which banks lend to each other overnight. It is the rate at which banks borrow from each other. The EONIA is the Euro Overnight Index Average. It is the average interest rate at which Eurozone banks lend to each other overnight.

The federal fund rate is virtually zero after the financial crisis of 2008. The EONIA rate is negative, which means that banks are paying each other to borrow money overnight.

- The reason why there is negative interest rate is that storing cash is costly. Banks have to pay for security, insurance, and other costs. So, they would rather lend money to each other at a negative interest rate than to store cash.

## Interest and Compound Interest

**Continuos Compounding**

Since the formula for copunding at a rate of r and assuming that the compunding happens n time a year, we have: 

$$A = P(1 + \frac{r}{n})^{nt}$$

If we assume that the compounding happens continuously, we have:

$$A = P e^{rt}$$

```python
def simple_interest(p, r, t):
    return p*(1+ r * t)

def compound_interest(p, r, t, n=1):
    return p * (1 + r/n) ** (n * t)

def compound_interest_n(p, r, t, n=1):
    return p * (1 + r/n) ** (n * t)

def continous_compound_interest(p, r, t):
    return p * np.exp(r * t)
```

```python
p = 1000
r = 0.05

t = np.arange(0, 51, 1)

t_cont = np.arange(0, 50.01, 0.01)
n = 3
fig = make_subplots(rows=1, cols=1)

fig.add_trace(go.Scatter(x=t, y=simple_interest(p, r, t), name='Simple Interest'), row=1, col=1)
fig.add_trace(go.Scatter(x=t, y=compound_interest(p, r, t), name='Compound Interest'), row=1, col=1)
fig.add_trace(go.Scatter(x=t, y=compound_interest_n(p, r, t, n), name=f'Compound Interest Compounded {n} times per year'), row=1, col=1)
fig.add_trace(go.Scatter(x=t_cont, y=continous_compound_interest(p, r, t_cont), name='Continous Compound Interest'), row=1, col=1)

fig.update_layout(title='Interest Rates', xaxis_title='Time', yaxis_title='Amount')

fig.show()
```

## Bonds

Bonds are a type of loan. They are a loan to a company or government. The company or government pays the bondholder a fixed amount of interest every year. The bondholder gets the principal back at the end of the loan period.

### Discount Bonds

A discount bond is a bond that is sold at a price less than its face value. The bondholder gets the face value of the bond at the end of the loan period.

Bond yields and bond prices have an inverse, or opposite, relationship. As interest rates increase, the price of a bond will decrease, and vice versa. A bond that offers bondholders a lower interest or coupon rate than the current market interest rate would likely be sold at a lower price than its face value. This lower price is due to the opportunity investors have to buy a similar bond or other securities that give a better return.

- "Yield" refers to the earnings generated and realized on an investment over a particular period of time. It's expressed as a percentage based on the invested amount, current market value, or face value of the security.

However, the "discount" in a discount bond doesn't necessarily mean that investors get a better yield than the market is offering. Instead, investors are getting a lower price to offset the bond's lower yield relative to interest rates in the current market. For example, if a corporate bond is trading at 980,itisconsideredadiscountbondsinceitsvalueisbelowthe1,000 par value. As a bond becomes discounted or decreases in price, it means its coupon rate is lower than current yields.

The maturity is given by: 

$$P = \frac{F}{(1 + \frac{r}{n})^{nt}}$$

where F is the face value, r is the interest rate, and t is the time to maturity.

If r is the rate of discount for discount bonds, the PDV (Present Discount Value) of a dollar, or any other amount, to be received in t periods is $$ P = \frac{1}{(1 + r)^t}$$

## Yield to Maturity

The yield to maturity is the rate of return on a bond if it is held until maturity. It is the discount rate that makes the present value of the bond equal to its face value. Yield to maturity is considered a long-term bond yield but is expressed as an annual rate.

The mathematical formula is:
$$
YTM = \left(\frac{F}{P}\right)^{\frac{1}{t}} - 1
$$

where $F$ is the face value, $P$ is the price, and $t$ is the time to maturity.

Another way to calculate this is:
$$
YTM = \frac{C + \frac{F-P}{t}}{\frac{F+P}{2}}
$$

where $C$ is the coupon rate.

## Interest Rates

### Nominal Interest Rate

Nominal interest rate refers to the interest rate before taking inflation into account. Nominal can also refer to the advertised or stated interest rate on a loan, without taking into account any fees or compounding of interest.

- The federal funds rate, the interest rate set by the Federal Reserve, is the short-term nominal interest rate that is the basis for other interest rates charged by banks and financial institutions.

### Effective Interest Rate

Effective interest rate is the interest rate that is actually paid on a loan, after taking into account compounding of interest and other fees. It is the actual rate of return on an investment, after taking into account compounding of interest and other fees.

We can calculate the effective interest rate using the following formula:

$$
\text{Effective Interest Rate} = \left(1 + \frac{r}{n}\right)^n - 1
$$

where $r$ is the nominal interest rate and $n$ is the number of compounding periods per year.

### Real Interest Rate

Real interest rate is the interest rate after taking inflation into account. It is the interest rate that is actually paid on a loan, after taking into account compounding of interest, other fees, and inflation.

We have:
$$
(1 + r_{real}) = \frac{1 + r_{nominal}}{1 + r_{inflation}}
$$
Which approximates to:
$$
r_{real} \approx r_{nominal} - r_{inflation}
$$

## A Summary of Interest Rates

The nominal interest rate, or coupon rate, is the actual price borrowers pay lenders, without accounting for any other economic factors.

The real interest rate accounts for inflation, giving a more precise reading of a borrower's buying power after the position has been redeemed.

The effective interest rate includes the impact of compounding, in which a bond might pay interest annually but compounds semiannually, increasing the overall return.


## Leverage

If a company or an individual borrows money to buy assets, we say that person or company is leveraging. Leveraging means you are putting more money into the asset than you have.

Leverage is the use of debt (borrowed capital) in order to undertake an investment or project. The result is to multiply the potential returns from a project. At the same time, leverage will also multiply the potential downside risk in case the investment does not pan out. When one refers to a company, property, or investment as "highly leveraged," it means that item has more debt than equity.

## Market Capitalization = Price per share * Number of shares outstanding

Market capitalization refers to the total dollar market value of a company's outstanding shares of stock. The investment community uses this figure to determine a company's size instead of sales or total asset figures. In an acquisition, the market cap is used to determine whether a takeover candidate represents a good value or not to the acquirer.

Companies are typically divided according to market capitalization: large-cap ($10 billion or more), mid-cap ($2 billion to $10 billion), and small-cap ($300 million to $2 billion).

The formula for market capitalization is: 

$$
\text{Market Capitalization} = \text{Price per share} \times \text{Number of shares outstanding}
$$

- Although it is used often to describe a company, the market cap does not measure the equity value of a company. Only a thorough analysis of a company's fundamentals can do that.

A **Non-profit** business is an organization whose key focus is to provide a charitable benefit to the community. ## Equity

- Equity represents the value that would be returned to a company’s shareholders if all of the assets were liquidated and all of the company's debts were paid off. We can also think of equity as a degree of residual ownership in a firm or asset after subtracting all debts associated with that asset.

A **Dividends** is a distribution of a portion of a company's earnings, decided by the board of directors, to a class of its shareholders. ## Dilution

The new shares dilute the value of the existing shares because the company now has more shares outstanding. The value of each share is reduced because the company has more shares outstanding.

## P/E Ratio

The **P/E ratio** is the ratio for valuing a company that measures its current share price relative to its per-share earnings. The price-to-earnings ratio is one of the most widely used ratios for valuing companies and is considered by many to be the best single measure of a stock's value. It is a measure of how much investors are willing to pay for each dollar of earnings.

Using Gordon Growth Model, we have:

$$ \frac{P}{E} = \frac{1}{r-g} $$

where r is the required rate of return (can be determined by CAPM model) and g is the growth rate.

If the stock covaries with the market, it has high r and thus low P/E.

If we know the earnings per share, P/E ratio can be used to determine the price per share. 

$$ \frac{P}{E} \text{ ratio} = \frac{\text{Market Value per Share}}{\text{Earnings per Share}} $$


## Pecking Order Theory

The **Pecking Order Theory** is a theory of capital structure that states that a firm will choose the least expensive source of funds available to finance its assets. The theory states that a firm will first use internally generated funds, then debt, and finally equity to finance its assets. The theory is based on the assumption that a firm's managers will choose the least expensive source of funds available to finance its assets.
